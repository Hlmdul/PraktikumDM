```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 2. Membaca file BMI
bmi_df = pd.read_csv('500_Person_Gender_Height_Weight_Index.csv')
bmi_df
```

```python
# 3. Menggunakan fungsi statistik untuk menghitung jumlah data, rata-rata, nilai maksimum, dan nilai minimum dari berat dan tinggi badan berdasarkan gender
print("\nStatistik Berat dan Tinggi Badan Berdasarkan Gender:")
stats = bmi_df.groupby('Gender').agg({
    'Height': ['count', 'mean', 'max', 'min'],
    'Weight': ['count', 'mean', 'max', 'min']
})
print(stats)
```

```python
# 4. Ubah kolom gender menjadi data biner: Female – 0, Male – 1
bmi_df['Gender_Binary'] = bmi_df['Gender'].apply(lambda x: 1 if x == 'Male' else 0)
print("\nData BMI dengan Gender Biner:")
print(bmi_df[['Gender', 'Gender_Binary']].head())
```

```python
# 5. Tampilkan grafik Index BMI berdasarkan gender
print("\nMenampilkan grafik Index BMI berdasarkan gender:")
bmi_df.groupby('Gender')['Index'].value_counts().unstack().plot(kind='bar', stacked=True)
plt.title('Distribusi BMI Index Berdasarkan Gender')
plt.xlabel('Gender')
plt.ylabel('Jumlah')
plt.show()
```
