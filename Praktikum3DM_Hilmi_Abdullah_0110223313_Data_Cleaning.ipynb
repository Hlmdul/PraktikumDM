```python
# Mengubah nama kolom untuk kemudahan mengakses data pada kolom
df2.rename(columns={
    'Jenis Kelamin': 'jenis_kelamin',
    'Tinggi (dalam cm)': 'tinggi',
    'Berat (dalam kg)': 'berat'
    }, inplace=True)
df2
```

Ditemukan 1 baris data yang mengandung Nan pada kolom berat, perlakuan selanjutnya adalah menghapus baris tersebut. penghapusan baris dilakukan atas pertimbangan 1 baris data yang mengandung nan tidak terlalu signifikan
secara proporsi terhadap keseluruhan data yang memiliki 76 baris data

```python
# menghapus baris yang mengandung nan
df2.dropna(subset=['berat'], inplace=True)
df2
```

```python
# Mengecek tipe data pada kolom berat dan tinggi
df2.info()
```

Dapat dilihat bahwa tipe data pada kolom tinggi dan berat belum berupa int atau float sehingga belum bisa diterapkan fungsi2 aritmatika. Untuk mengubah tipe data tersebut akan dilakukan method .to_numeric pada kolom tinggi dan berat, namun sebelum itu dapat dilihat bahwa terdapat inkonsistensi dalam penulisan data tinggi dan berat seperti adanya tambahan satuan kg pada berat, dan ada data yang menggunakan tanda baca koma (,) untuk pemisah digit desimal, bila langsung dipanggil method .to_numeric, data2 tersebut akan hilang sehingga dilakukan beberapa method .str.replace() untuk mengatasi inkonsistensi tersebut

```python
# Menghilangkan Inkonsistensi penulisan data
df2['berat'] = df2['berat'].str.replace(' kg', '', regex=False)
df2['berat'] = df2['berat'].str.replace('kg', '', regex=False)
df2['berat'] = df2['berat'].str.replace(',', '.', regex=False)

df2['tinggi'] = df2['tinggi'].str.replace(' cm', '', regex=False)
df2['tinggi'] = df2['tinggi'].str.replace('cm', '', regex=False)
df2['tinggi'] = df2['tinggi'].str.replace(',', '.', regex=False)
df2
```

```python
# Mengubah tipe data pada kolom tinggi dan berat menjadi numeric
df2['tinggi'] = pd.to_numeric(df2['tinggi'], errors='coerce')
df2['berat'] = pd.to_numeric(df2['berat'], errors='coerce')
df2.info()
```

Ditemukan outlier pada boxplot di atas, maka dilakukan investigasi mengenai posisi outlier dan apakah data outlier tersebut valid atau ada kesalahan input

```python
q1_tinggi = df2['tinggi'].quantile(0.25)
q3_tinggi = df2['tinggi'].quantile(0.75)
iqr_tinggi = q3_tinggi - q1_tinggi

batas_atas_tinggi = q3_tinggi + 1.5 * iqr_tinggi
batas_bawah_tinggi = q1_tinggi - 1.5 * iqr_tinggi

outlier_tinggi_index = df2[(df2['tinggi'] > batas_atas_tinggi) | (df2['tinggi'] < batas_bawah_tinggi)]


q1_berat = df2['berat'].quantile(0.25)
q3_berat = df2['berat'].quantile(0.75)
iqr_berat = q3_berat - q1_berat
batas_atas_berat = q3_berat + 1.5 * iqr_berat
batas_bawah_berat = q1_berat - 1.5 * iqr_berat
outlier_berat_index = df2[(df2['berat'] > batas_atas_berat) | (df2['berat'] < batas_bawah_berat)]

# Tampilkan posisi outlier
print("Posisi outlier pada kolom 'tinggi': \n", outlier_tinggi_index)
print("\nPosisi outlier pada kolom 'berat': \n", outlier_berat_index)
print(batas_atas_tinggi,q3_tinggi, q1_tinggi, batas_bawah_tinggi)
print(batas_atas_berat,q3_berat, q1_berat, batas_bawah_berat)
```

```python
# Membuat Variabel baru BMI
df2['BMI'] = df2['berat'] / (df2['tinggi']/100)**2
df2
```

Selanjutnya berdasarkan variabel baru BMI akan dilakukan Grouping dengan klasifikasi sebagai berikut:


*   underweight (<18.5)
*   normal (18.5-24.9)
*   overweight (25.0-29.9)
*   obesity (>29.9)



```python
# Definisikan fungsi untuk klasifikasi BMI
def klasifikasi_bmi(bmi):
    if bmi < 18.5:
        return 'underweight'
    elif 18.5 <= bmi <= 24.9:
        return 'normal'
    elif 25.0 <= bmi <= 29.9:
        return 'overweight'
    else:
        return 'obesity'

# menerapkann fungsi ke kolom 'BMI' dan buat kolom baru 'klasifikasi_bmi'
df2['klasifikasi_bmi'] = df2['BMI'].apply(klasifikasi_bmi)

# grouping berdasarkan 'klasifikasi_bmi'
grouped_bmi = df2.groupby('klasifikasi_bmi')

# menampilkan agregat data berdasarkan grouping bmi
print(grouped_bmi.describe())

print('\nRata-rata tinggi badan berdasarkan klasifikasi BMI')
print(grouped_bmi['tinggi'].mean())
```

```python
# Grouping berdasarkan Jenis Kelamin
grouped_jenis_kelamin = df2.groupby(['jenis_kelamin'])
grouped_jenis_kelamin.describe()
```

```python
# menampilkan rata-rata berat dan tinggi dari tiap jenis kelamin
print('Rata-rata tinggi dan berat berdasarkan jenis kelamin: ')
print(grouped_jenis_kelamin[['tinggi', 'berat']].mean())
```

Terakhir akan dilakukan 3 tipe Normalisasi data yaitu
1. Min-Max Normalization
2. Z-Score Normalization
3. Scaling nilai desimal


```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# Min-Max Normalization

# Inisialisasi MinMaxScaler
scaler = MinMaxScaler()

# Lakukan Min-Max Normalization untuk kolom 'tinggi' dan 'berat'
df2[['tinggi_minmax', 'berat_minmax']] = scaler.fit_transform(df2[['tinggi', 'berat']])

# Menampilkan hasil normalisasi minmax
df2[['tinggi','tinggi_minmax', 'berat', 'berat_minmax']]
```

```python
# Z-Score

# Inisialisasi StandardScaler
scaler = StandardScaler()

# Lakukan Z-Score Normalization untuk kolom 'tinggi' dan 'berat'
df2[['tinggi_zscore', 'berat_zscore']] = scaler.fit_transform(df2[['tinggi', 'berat']])

# Menampilkan hasil normalisasi zscore
df2[['tinggi','tinggi_zscore', 'berat', 'berat_zscore']]
```

```python
# Scaling Factor
scaling_factor_tinggi = 1000
scaling_factor_berat = 100

# Lakukan scaling nilai desimal untuk kolom 'tinggi' dan 'berat'
df2['tinggi_decimal'] = df2['tinggi'] / scaling_factor_tinggi
df2['berat_decimal'] = df2['berat'] / scaling_factor_berat

# Menampilkan hasil scaling nilai desimal
df2[['tinggi', 'tinggi_decimal', 'berat', 'berat_decimal']]
```
